```python
#we must first clean the data. Check for nulls and duplications
#Check for null values
kepler.isna().sum()
```

```python
rem_col = ['koi_srad_err2', 'koi_srad_err1', 'koi_slogg_err2', 'koi_slogg_err1', 'koi_steff_err2', 'koi_steff_err1', 
           'koi_insol_err2', 'koi_insol_err1', 'koi_teq_err2', 'koi_teq_err1', 'koi_prad_err1', 'koi_prad_err2', 'koi_depth_err2'
          , 'koi_depth_err1', 'kepler_name', 'koi_period_err1', 'koi_period_err2', 'koi_time0bk_err1', 'koi_time0bk_err2', 'koi_impact_err1'
          , 'koi_impact_err2', 'koi_duration_err1', 'koi_duration_err2', 'koi_depth_err1', 'koi_depth_err2']

kep = kepler.drop(rem_col, axis = 1)
kep.shape
```

```python
keplerdf = kep.dropna()
print(keplerdf.shape)
keplerdf.isna().sum()
```

```python
#are there duplicates in the dataset? 
keplerdf.duplicated().sum()
```

```python
#No duplicates, hence we may carry on and do statistical analysis on it.
corr1_matrix = keplerdf.corr()
plt.figure(figsize = (20,15))
sns.heatmap(corr1_matrix, annot = True, cmap = 'coolwarm', linewidths=.5)
plt.title('Correlation Matrix Heatmap')
plt.show()
```

```python
plt.figure(figsize=(15, 18))

for i in range(0, len(keplerdf.columns)):
    plt.subplot(10, 4, i+1)
    plt.hist(keplerdf[keplerdf.columns[i]], density = True, bins = 5, align='mid')
    plt.xlabel(keplerdf.columns[i])
```
